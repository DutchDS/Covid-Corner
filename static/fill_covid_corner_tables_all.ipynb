```python
import pandas as pd
import os
from datetime import datetime, timedelta
from sqlalchemy import create_engine
```

```python
# Set where .csv files needs to be saved
base_output_path = "../../data"

# Create a POSTGRES database with the name 'COVID19_db'
# Replace username:password if it's not set to postgres:postgres
DATABASE_URI = os.environ.get('DATABASE_URL', '') or "postgresql://postgres:postgres@localhost:5432/COVID19_db"
print(DATABASE_URI)

engine = create_engine(DATABASE_URI)
```

```python
url_categories = os.path.join(base_output_path, 'categories.csv')
url_addresses = os.path.join(base_output_path, 'addresses.csv')
```

```python
df_categories = pd.read_csv(url_categories)
df_addresses = pd.read_csv(url_addresses)
```

```python
df_categories
df_addresses
```

```python
# engine.execute( '''TRUNCATE TABLE categories''' )
df_categories.to_sql(name='categories', con=engine, if_exists='append', index=False)

# engine.execute( '''TRUNCATE TABLE addresses''' )
df_addresses.to_sql(name='addresses', con=engine, if_exists='append', index=False)
```

```python
# Find last time covid_data was loaded
query_str = open('sql/max_db_date.sql')
query_text = ""
for text in query_str:
    query_text = query_text + text

rs = engine.execute(query_text)

for i in rs:
    last_db_date = (i[0])
print(last_db_date)
```
